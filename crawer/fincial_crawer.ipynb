```python
import requests 
from bs4 import BeautifulSoup
target_url = "https://money.udn.com/money/index"
rs = requests.session()
res = rs.get(target_url, verify=False)
soup = BeautifulSoup(res.text, 'html.parser')
content = ""
print(soup)
```

```python
for data in soup.select('div.only_mobile a'):
    title = data.text
    link =data['href']
    content += "{}\n{}\n\n".format(title,link)
    print(title)
```
