```python
import pandas
import requests 
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
import re

target_url = "http://www.tse.com.tw/zh/news/newsList"
print('Start parsing  TSE....')
rs = requests.session()
res = rs.get(target_url, verify=False)
soup = BeautifulSoup(res.text, 'html.parser')
content = ""
print(soup)

```

```python
for data in soup.select("td a"):
    title = data.text
    head="http://www.tse.com.tw"
    link=head+data['href']
    detail ='{}\n{}\n\n'.format(title, link)
    print(type(detail))
```

```python
content +="{}\n\n{}".format(title,link)
print(content)
```
