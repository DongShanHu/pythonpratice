```python
import requests 
from bs4 import BeautifulSoup
# https://nba.hupu.com/games/2018-11-11
target_url = "https://nba.hupu.com/games"
rs = requests.session()
res = rs.get(target_url, verify=False)
soup = BeautifulSoup(res.text, 'html.parser')
content = ""
print(soup)
```

```python
def hupu_news():
    content ="24小時新聞排行榜\n\n"
    for news_tilte in soup.select('div.bbs_a ul.hour a'):
        title=news_tilte.text
        link=news_tilte['href']
        content +="{}\n{}\n\n".format(title,link)
    print(content)
hupu_news()
```

```python
def box_score():
    content=""
    for box in soup.select('div.gamecenter_content_l div.team_vs_a'):
        title=box.text
        title=title.strip()
        content+="{}\n{}\n".format(title,',')
        print(content)
#         link=box['href']
#         content +="{}\n{}\n\n".format(title,link)     
box_score()
```

```python
def team_rank():
    target_url = "https://nba.hupu.com/standings"
    rs = requests.session()
    res = rs.get(target_url, verify=False)
    soup = BeautifulSoup(res.text, 'html.parser')
    content = ""
    for a in soup.select('div.rank_data tr td.left a'):
        title=a.text
        link=a['href']
        content +="{}\n{}\n\n".format(title,link)    
        print(title)
    print(content)
team_rank()
```
