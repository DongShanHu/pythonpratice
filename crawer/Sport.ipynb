```python
import pandas
import requests 
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
import re
def playsport():
    target_url = "https://www.playsport.cc/livescore.php?aid=1"
    rs = requests.session()
    res = rs.get(target_url, verify=False)
    soup = BeautifulSoup(res.text, 'html.parser')
    return soup
def teams():
    soup=playsport()
    b=[]
    for team in soup.select("div.st table.scorebox th"):
        a = team.text
        names=a
        b.append(names)
    return b
def scores():
    soup=playsport()
    a=[]
    for data2 in soup.select("td.big_score"):
        b = data2.text
        a.append(b)
    return a
def mlb():
    print('Start parsing  MLB....')
    my_list = [teams(),scores()]
    content="隊名：{0[0]}, 分數 {0[1]}".format(my_list)
    return content
#     print("隊名：{0[0]}, 分數 {0[1]}".format(my_list))  # "0" 是必须的
```
